```python
import pandas as pd
import os
```

```python
file_path = "D:/Kerner-Lab/geo-clip-paper-implementation/dataset/yfcc4k/yfcc4k.txt"
```

```python
## read the csv file where I don't have the number of columns in the entire file
# read .txt file
df = pd.read_csv(file_path, delimiter="\t", header=None)
```

```python
# show the columns of the dataframe 9-12
df.head()
```

```python
# keep columns 1, 12, 13
df = df.iloc[:, [1, 12, 13]]
```

```python
#rename columns to id, longitude, latitude
df = df.rename(columns={1: 'id', 12: 'longitude', 13: 'latitude'})
```

```python
df
```

```python
file_paths = []
file_ids = []

for r, d, f in os.walk('D:/Kerner-Lab/geo-clip-paper-implementation/dataset/yfcc4k/yfcc4k/'):
    for file in f:
        if '.jpg' in file:
            print(os.path.join(r, file))
            print(os.path.getsize(os.path.join(r, file)))
            if os.path.getsize(os.path.join(r, file)) == 0:
                os.remove(os.path.join(r, file))
            else:
                file_paths.append(os.path.join(r, file))
                filename = file.split('/')[-1]
                file_ids.append(filename.split('.')[0])

print(file_paths)
print(file_ids)
```

```python
file_df = pd.DataFrame(list(zip(file_paths, file_ids)), columns = ['file_path', 'file_id'])
```

```python
file_df
```

```python
file_df['file_id'] = file_df['file_id'].astype(str)
df['id'] = df['id'].astype(str)
```

```python
modified_df = pd.merge(df, file_df, how = 'inner', left_on = 'id', right_on = 'file_id')
```

```python
modified_df
```

```python
import matplotlib.pyplot as plt
```

```python
#plot latlong data
plt.scatter(modified_df['longitude'], modified_df['latitude'], c='red', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Photo Locations')
plt.show()
```

```python
modified_df = modified_df.drop(columns=['file_id'])
```

```python
modified_df.to_csv('D:/Kerner-Lab/geo-clip-paper-implementation/dataset/yfcc4k/yfcc4k.csv', index=False)
```
